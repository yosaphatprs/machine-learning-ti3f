# 1. Buatlah model klasifikasi dengan menggunakan SVM untuk data suara, voice.csv. 

```python
# import library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data dari local

df = pd.read_csv('voice.csv')
df.head()
```

```python
# Analisis data diatas

df.info()
```

```python
# proses encoding untuk mengubah data menjadi numerik

from sklearn.preprocessing import StandardScaler, LabelEncoder

le = LabelEncoder()
df['label'] = le.fit_transform(df['label'])
df.head()
```

```python
# membuat model SVM nya

from sklearn.svm import SVC

svm_classifier = SVC(kernel='linear')

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop(['label'],axis=1))
scaled_features
```

```python
# melakukan split data menjadi perbandingan 80:20

from sklearn.model_selection import train_test_split

X = scaled_features
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1000)

svm_classifier.fit(X_train,y_train)
```

```python
# Melakukan prediksi data

from sklearn.metrics import accuracy_score, classification_report

# Make predictions on the test set
y_pred = svm_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate a classification report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

# 2. Buatlah model klasfikasi Multinomial Naive Bayes dengan ketentuan,
## a. Menggunakan data spam.csv
## b. Fitur CountVectorizer dengan mengaktifkan stop_words
## c. Evaluasi hasilnya

```python
# membaca data dari spam.csv

df_spam =  pd.read_csv('spam.csv', encoding='latin-1')
df_spam.head(10)
```

```python
df_spam.info()
df_spam.describe()
```

```python
# pemilahan data yang tidak digunakan

df_spam.drop(df_spam.iloc[:,2:],axis=1,inplace=True)
df_spam.head()
```

```python
# memisah data untuk training dan untuk sampling

X = df_spam['v2']
y = df_spam['v1']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
# transformasi dengan CountVectorized 

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english')

X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
# membuat model naive bayes

from sklearn.naive_bayes import MultinomialNB

nb_classifier = MultinomialNB()

nb_classifier.fit(X_train,y_train)
```

```python
# melakukan prediksi pada data

y_pred = nb_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

# 3. Buatlah model klasfikasi Multinomial Naive Bayes dengan ketentuan,
## a. Menggunakan data spam.csv
## b. Fitur TF-IDF dengan mengaktifkan stop_words
## c. Evaluasi hasilnya dan bandingkan dengan hasil pada Tugas no 2.
## d. Berikan kesimpulan fitur mana yang terbaik pada kasus data spam.csv

```python
# membaca data 

data_spam = pd.read_csv('spam.csv', encoding='latin-1')
data_spam.head(10)
```

```python
data_spam.info()
data_spam.describe()
```

```python
# pemilahan data yang tidak digunakan

data_spam.drop(data_spam.iloc[:,2:],axis=1,inplace=True)
data_spam.head()
```

```python
# memisah antara data untuk training dan sampling

X = data_spam['v2']
y = data_spam['v1']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
# mentranformasi data menggunakan TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
```

```python
# membuat mode naive bayes

from sklearn.naive_bayes import MultinomialNB 

nb_classifier = MultinomialNB()

nb_classifier.fit(X_train,y_train)
```

```python
# melakukan prediksi dari data yang sudah di proses

y_pred = nb_classifier.predict(X_test)

tf_accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", tf_accuracy)

# Generate a classification report
tf_report = classification_report(y_test, y_pred)
print("Classification Report:\n", tf_report)
```
