#### Nomor 1

```python
import pandas as pd
df = pd.read_csv('../data/voice.csv')
df.head()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

model = SVC(kernel='rbf')
x = df.drop(['label'], axis=1)
y = df.label.values

std = StandardScaler()
x = std.fit_transform(x)

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=30)
model.fit(X_train, y_train)

y_train_predic = model.predict(X_train)
acc_train_svm = accuracy_score(y_train, y_train_predic)

y_test_predic = model.predict(X_test)
acc_test_svm = accuracy_score(y_test, y_test_predic)

print(f'Hasil akurasi prediksi data train: {acc_train_svm}')
print(f'Hasil akurasi prediksi data test: {acc_test_svm}')
```

#### Nomor 2

```python
data = pd.read_csv('../data/spam.csv', encoding='latin-1')
data = data.drop(data.iloc[:,2:], axis=1)

new_cols = {
    'v1': 'Labels',
    'v2': 'SMS'
}
data = data.rename(columns=new_cols)

new_labels = {
    'spam': 1,
    'ham': 0
}
data['Labels'] = data['Labels'].map(new_labels)

data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

X = data['SMS'].values 
y = data['Labels'].values 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
bow = CountVectorizer(stop_words='english') # memberikan stop_words untuk menghapus kata yang tidak bermakna
X_train = bow.fit_transform(X_train)
X_test = bow.transform(X_test)

mnb = MultinomialNB()
mnb.fit(X_train, y_train)

y_pred_train = mnb.predict(X_train)
acc_train_vectorizer = accuracy_score(y_train, y_pred_train)

y_pred_test = mnb.predict(X_test)
acc_test_vectorizer = accuracy_score(y_test, y_pred_test)

print(f'Hasil akurasi data train: {acc_train_vectorizer}')
print(f'Hasil akurasi data test: {acc_test_vectorizer}')
```

#### Nomor 3

```python
data2 = pd.read_csv('../data/spam.csv', encoding='latin-1')
data2 = data2.drop(data2.iloc[:,2:], axis=1)

new_cols = {
    'v1': 'Labels',
    'v2': 'SMS'
}
data2 = data2.rename(columns=new_cols)

new_labels = {
    'spam': 1,
    'ham': 0
}
data2['Labels'] = data2['Labels'].map(new_labels)

data2.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfid = TfidfVectorizer(stop_words='english')

X = data2['SMS'].values 
y = data2['Labels'].values 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
X_train = tfid.fit_transform(X_train)
X_test = tfid.transform(X_test)

mnb2 = MultinomialNB()
mnb2.fit(X_train, y_train)

y_pred_train = mnb2.predict(X_train)
acc_train_tfid = accuracy_score(y_train, y_pred_train)

y_pred_test = mnb2.predict(X_test)
acc_test_tfid = accuracy_score(y_test, y_pred_test)

print(f'Hasil akurasi data train: {acc_train_tfid}')
print(f'Hasil akurasi data test: {acc_test_tfid}')
```

##### Perbandingan dan Kesimpulan

```python
print(f'Hasil akurasi data train menggunakan CountVectorizer: {acc_train_vectorizer}')
print(f'Hasil akurasi data test menggunakan CountVectorizer: {acc_test_vectorizer}')
print(f'Hasil akurasi data train menggunakan TF-IDF: {acc_train_tfid}')
print(f'Hasil akurasi data test menggunakan TF-IDF: {acc_test_tfid}')
```

Kesimpulan : Jika data text ditransformasi menjadi token menggunakan fitur CountVectorizer, maka training model menggunakan data tersebut akan menghasilkan keakurasian yang lebih tinggi dibandingkan dengan data yang ditransformasi menggunakan fitur TF-IDF
