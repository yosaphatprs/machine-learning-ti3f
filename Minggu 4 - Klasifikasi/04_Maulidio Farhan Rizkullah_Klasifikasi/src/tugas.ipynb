### Nomor 1

```python
# import library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../docs/voice.csv')

df.head()
```

```python
new_labels = {
    'male' : 1,
    'female' : 0
}

# Encode label
df['label'] = df['label'].map(new_labels)

# Cek data
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Memisahkan fitur dengan label menggunakan X sebagai fitur dan y sebagai label
X = df.iloc[:, :20].values
y = df['label'].values

std = StandardScaler()  # membuat objek dari StandartScaler
X = std.fit_transform(X)

# split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size=0.2, random_state=1)
df.info()

```

```python
df.head()

```

```python
# import fungsi dari library sklearn
from sklearn import metrics
from sklearn.svm import SVC

# inisiasi SVM
svc=SVC()

svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
acc = metrics.accuracy_score(y_test,y_pred)
print(f'Accuracy Score : {int(acc * 100)}%')
```

### Nomor 2

```python
from sklearn.feature_extraction.text import CountVectorizer

df = pd.read_csv('../docs/spam.csv', encoding='latin-1', usecols=["v1", "v2"])
df.columns = ["labels", "message"]

df.info()
```

```python
new_labels = {
    'spam': 1,
    'ham': 0
}

# encode label
df['labels'] = df['labels'].map(new_labels)
df.head()
```

```python
X = df["message"]
y = df["labels"]
```

Fitur CountVectorizer dengan mengaktifkan stop_words

```python
# split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 50, test_size = 0.3)

cv = CountVectorizer(stop_words = 'english')

X_train = cv.fit_transform(X_train)
X_test = cv.transform(X_test)
```

```python
print(len(cv.get_feature_names_out()))
print(f'Dimensi data: {X_train.shape}')
```

```python
from sklearn.naive_bayes import MultinomialNB

# Inisiasi MultinomialNB
mnb = MultinomialNB()

mnb.fit(X_train, y_train)

y_pred_train_mnb = mnb.predict(X_train)

acc_train_mnb = metrics.accuracy_score(y_train, y_pred_train_mnb)

y_pred_test_mnb = mnb.predict(X_test)

# Evaluasi akurasi data training
acc_test_mnb = metrics.accuracy_score(y_test, y_pred_test_mnb)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train_mnb}')
print(f'Hasil akurasi data test: {acc_test_mnb}')
```

### Nomor 3

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)

# Inisiasi TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

X_train = tfidf.fit_transform(X_train)

X_test = tfidf.transform(X_test)
```

```python
print(len(tfidf.get_feature_names_out()))
print(f'Dimensi data: {X_train.shape}')
```

```python
from sklearn.naive_bayes import MultinomialNB 

# Inisiasi MultinomialNB
mnb = MultinomialNB()

mnb.fit(X_train, y_train)

y_pred_train_mnb = mnb.predict(X_train)

train_acc_mnb = metrics.accuracy_score(y_train, y_pred_train_mnb)

y_pred_test_mnb = mnb.predict(X_test)

# Evaluasi akurasi data training
test_acc_mnb = metrics.accuracy_score(y_test, y_pred_test_mnb)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {train_acc_mnb}')
print(f'Hasil akurasi data test: {test_acc_mnb}')
```

### Kesimpulan

Berdasarkan hasil akurasi data train dan data test pada model klasifikasi Multinomial Naive Bayes yang menggunakan 2 fitur, yaitu CountVectorizer dengan mengaktifkan stop_words dan TF-IDF dengan mengaktifkan stop_words, dapat disimpulkan bahwa fitur terbaik pada kasus data spam.csv adalah CountVectorizer karena menghasilkan akurasi yang lebih baik dibandingkan TF-IDF
