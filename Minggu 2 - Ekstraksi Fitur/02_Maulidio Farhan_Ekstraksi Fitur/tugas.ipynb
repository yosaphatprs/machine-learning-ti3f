## Nomor 1

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler


dpath = 'wbc.csv'
df = pd.read_csv(dpath)
df.head()
```

```python
df.info()
```

```python
df = df.drop(['id','Unnamed: 32'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

# Nomor 2

```python
le = LabelEncoder() 
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df.head(20)
```

# Nomor 3

```python
# memilih kolom yang memiliki nilai numeric
numeric = df.select_dtypes(include=['int64', 'float64'])
    
std = StandardScaler() #membuat objek dari StandartScaler
df[numeric.columns] = std.fit_transform(numeric) #untuk melakukan standarisasi pada numeric data
    
    
    
```

```python
df.head()
```

# Nomor 4

```python
from sklearn.model_selection import train_test_split #import library train_test_split 


df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis']) #melakukan split data

# Memeriksa masing-masing ukuran data

print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
```
