```python
import pandas as pd #impor library pandas
data = 'wbc.csv'
#dppath disini menyimpan path data kita yg nanti akan dibaca
df = pd.read_csv('D:\SEMS 5\MACHINE LEARNING\wbc.csv') #membaca path yg sudah kita tulis diatas
df.head() #menampilkan lima baris pertama dari DataFrame tersebut.
```

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
#cek info datanya
df.info()
```

```python
# Karena terdapat 1 variabel yang null yaitu unnamed: 32 dan variabel id, maka kita akan menyingkirkan variabel tersebut
df = df.drop(columns=['id','Unnamed: 32'])
df.head()
```

```python
# misahin (numeric) yg bisa dipakai dan yang tidak bisa
numeric_columns = df.select_dtypes(include=['number'])
non_numeric_columns = df.select_dtypes(exclude=['number'])

# nampilin baris pertama dari DataFrame yg numerik
print("Data kolom numerik:")
print(numeric_columns.head())

# nampilin yg tidak bisa
print("Variabel yang tidak bisa dipakai:")
print(non_numeric_columns)


```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
from sklearn.preprocessing import LabelEncoder #impor dri sklearn
le = LabelEncoder() #membuat objek LabelEncoder
df['diagnosis']=le.fit_transform(df['diagnosis']) #proses encode nya
df.head()
```

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler() #inisialisasi objek std

num_variabel = [  #daftar kolom numerik yg akan distandarisasi
    'radius_mean',
    'texture_mean',
    'perimeter_mean',
    'area_mean',
    'smoothness_mean',
    'compactness_mean',
    'concavity_mean',
    'concave points_mean',
    'symmetry_mean',
    'fractal_dimension_mean',
    'radius_se',
    'texture_se',
    'perimeter_se',
    'area_se',
    'smoothness_se',
    'compactness_se',
    'concavity_se',
    'concave points_se',
    'symmetry_se',
    'fractal_dimension_se',
    'radius_worst',
    'texture_worst',
    'perimeter_worst',
    'area_worst',
    'smoothness_worst',
    'compactness_worst',
    'concavity_worst',
    'concave points_worst',
    'symmetry_worst',
    'fractal_dimension_worst'
]
df[num_variabel] = std.fit_transform(df[num_variabel])
df.head()
```

4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
from sklearn.model_selection import train_test_split #impor fungsi splitnya

df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])
#membagi data awal menjadi dua bagian: df_train dan df_unseen dengan rasio 80:20.
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0, stratify=df_unseen['diagnosis'])
#mmebagi data 

print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data val:\n{df_val.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
```
