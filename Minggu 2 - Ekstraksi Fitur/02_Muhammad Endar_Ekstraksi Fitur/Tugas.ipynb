Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,

2. Lakukan proses encoding pada kolom "diagnosis".

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

## Jawaban

1. Import Libraries: Impor library yang diperlukan seperti pandas, numpy, sklearn, dan lainnya.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

2. Load Data: Muat data "Wisconsin Breast Cancer" ke dalam Jupyter Notebook.

```python
data = pd.read_csv('docs/wbc.csv')
```

```python
data.head()
```

3. Pemisahan Variable : Pisahkan variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
columns_to_exclude = ['Unnamed: 32', 'id']
data_filtered = data.drop(columns=columns_to_exclude)
```

```python
data_filtered.head()
```

4. Encoding: Lakukan proses encoding pada kolom "diagnosis" menggunakan LabelEncoder.

```python
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])
```

```python
data.head()
```

5. Standarisasi : Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik, kecuali kolom target.

```python
scaler = StandardScaler()
numeric_columns = data_filtered.columns.difference(['diagnosis'])
data_filtered[numeric_columns] = scaler.fit_transform(data_filtered[numeric_columns])
data_filtered.head()
```

6. Stratified Split: Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
X = data_filtered.drop('diagnosis', axis=1)  
y = data_filtered['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print("Jumlah data latih (X_train):", len(X_train))
print("Jumlah data uji (X_test):", len(X_test))
```
