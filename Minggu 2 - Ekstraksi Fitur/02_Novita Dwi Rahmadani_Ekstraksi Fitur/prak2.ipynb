```python
#Langkah 0 - Load Library
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
#Langkah - Load Data
dpath = 'data/Titanic-Dataset-fixed.csv' #mendefinisikan yang berisi path/lokasi file csv yg ingin dibaca
df = pd.read_csv(dpath) #untuk membaca file CSV yang terletak di path yang telah didefinisikan sebelumnya
df.head() # membaca file CSV dan memuat datanya ke dalam DataFrame
```

```python
#Langkah 2 - Slice Data
df = df[['Survived', 'Pclass', 'Age', 'Sex', 'Cabin']] 3aris ini memilih kolom-kolom 'Survived', 'Pclass', 'Age', 'Sex', dan 'Cabin' dari DataFrame df. Ini menghasilkan DataFrame baru yang hanya berisi kolom-kolom ini.
df.head() #menampilkan kelima data yang sudah difilter
```

```python
#Langkah 3 - Encoding
le = LabelEncoder() # membuat objek dari LabelEncoder
df['Sex'] = le.fit_transform(df['Sex']) # proses encoding
df['Cabin'] = le.fit_transform(df['Cabin']) # proses encoding

#Label encoding digunakan untuk mengubah nilai-nilai dalam kolom kategori (categorical) menjadi bilangan bulat (integer) agar bisa digunakan dalam berbagai algoritma machine learning yang memerlukan input berupa bilangan bulat
```

```python
#Langkah 4 - Verifikasi Hasil
df.head() #menampilkan beberapa data teratasa dari data frame
```

```python
#Langkah 5 - Standarisasi
std = StandardScaler() #menciptakan sebuah objek dari kelas StandardScaler yang akan digunakan untuk melakukan standarisasi.
df['Age'] = std.fit_transform(df[['Age']]) #mengambil kolom age untuk mengubah nilai dari kolom age menjadinilai yg telah distandarisasi.
#Standarisasi dalam hal ini mengubah nilai-nilai dalam kolom tersebut sehingga memiliki rata-rata 0 dan deviasi standar 1. Hasil standarisasi kemudian disimpan kembali dalam kolom 'Age' dalam DataFrame df.
```
