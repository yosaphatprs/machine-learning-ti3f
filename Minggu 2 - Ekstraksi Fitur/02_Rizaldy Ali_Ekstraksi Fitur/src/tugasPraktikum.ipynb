**Mengolah Data**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
df = pd.read_csv('../data/wbc.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

**1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan**

```python
kolom_kec = ['id', 'Unnamed: 32']
df = df.drop(columns=kolom_kec)
```

```python
df.head()
```

#### Penjelasan

Kode diatas yaitu menghapus kolom 'id' dan kolom 'Unnamed: 32" dari df dan kemudian menugaskan df yang telah dimodifikasi ke variabel 'df' kembali. Hal ini berarti setelah kode dijalankan menggunakan perintah 'df.head()' maka tidak akan ada lagi dua kolom tersebut

**2. Lakukan proses encoding pada kolom "diagnosis".**

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
df.head()
```

#### Penjelasan
Kode diatas memiliki arti dimana LabelEncoder menugaskan ke variabel le yang digunakan untuk mengubah label atau ketegori dalam data bilangan bulat. Lalu ada kode df['diagnosis'] = le.fit_transform(df['diagnosis']) yang berfungsi mengubah nilai dalam kolom diagnosis menjadi bilangan bulat. Dan kode terakhir yaitu df.head() yang berguna untuk mencetak beberapa baris teratas dari df

**3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.**

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
kolom_numerik = df.columns.difference(['diagnosis'])
df[kolom_numerik] = std.fit_transform(df[kolom_numerik])
df.head()
```

#### Penjelasan
Kode diatas berisi perintah dimana kode kolom_numerik = df.columns.difference(['diagnosis']) berfungsi untuk mengidentifikasi kolom-kolom numerik dalam df. Kolom ini akan diambil dengan menghilangan kolom diagnosis dan akan berisi daftar kolom yang berisi numerik. Lalu perintah df[kolom_numerik] = std.fit_transform(df[kolom_numerik]) digunakan untuk menghitung rata-rata dan standarisasi yang hasilnya akan disimpan pada df. Perintah df.head() berguna untuk mencetak beberapa baris pertama dalam df yang sudah di standarisasi

**4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.**

```python
# Split data
from sklearn.model_selection import train_test_split

# Rasio yang akan kita gunakan adalah 80:20
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Cek masing-masing ukuran data

print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data latih:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data uji:\n{df_test.diagnosis.value_counts()}')
print(f'Jumlah data latih: ',len(df_train))
print(f'Jumlah data uji: ',len(df_test))
```

#### Penjelasan
Kode diatas berfungsi untuk melakukan operasi terkait pemisahan data menjadi data latih dan data uji dengan rasio 80:20

- df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis']) kode ini berfungsi memanggil fungsi untuk membagi df menjadi dua bagian yaitu data latih dan data uji dengan rasio pengambilan data uji sebesar 20%
- print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}') kode ini mencetak data asli yang belum dibagi
- print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}') kode ini mencetak jumlah data latih setelah dipisah
- print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}') kode ini mencetak jumlah data uji setelah dipisah
