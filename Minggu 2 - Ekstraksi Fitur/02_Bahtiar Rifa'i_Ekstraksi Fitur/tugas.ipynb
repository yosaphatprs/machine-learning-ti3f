Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = 'data/wbc.csv'
df = pd.read_csv(data)
df.head()
```

```python
# Variable yang dapat digunakan adalah semua kecuali "id" dan "Unnamed"

# Pilih semua kolom kecuali "id" dan "Unnamed"
df = df[df.columns[~df.columns.isin(['id', 'Unnamed: 32'])]]

# Menampilkan lima baris pertama dari DataFrame yang sudah dipilih kolomnya
df.head()
```

Lakukan proses encoding pada kolom "diagnosis".

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding
```

```python
df.head() # Tampilkan lima baris pertama dari DataFrame setelah standarisasi
```

Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
# Pilih kolom-kolom yang memiliki nilai numerik (kecuali "diagnosis")
numeric_columns = df.columns[1:]

# Inisialisasi objek StandardScaler
scaler = StandardScaler()

# Lakukan standarisasi pada kolom-kolom numerik
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# Tampilkan lima baris pertama dari DataFrame setelah standarisasi
df.head()
```

Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
# Memisahkan dataset menjadi training set dan df_unseen (80% training, 20% df_unseen)
df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Memisahkan df_unseen menjadi validation set dan test set (masing-masing 50% dari df_unseen)
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0, stratify=df_unseen['diagnosis'])

# Menampilkan jumlah label kelas dalam setiap bagian dataset
print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data val:\n{df_val.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
```
