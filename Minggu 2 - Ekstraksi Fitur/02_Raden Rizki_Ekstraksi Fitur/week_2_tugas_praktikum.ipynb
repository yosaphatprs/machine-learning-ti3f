```python
import pandas as pd # Import Library Pandas  dan memberi alias sebagai pd
from sklearn.preprocessing import LabelEncoder, StandardScaler # Import Library LabelEncoder dan StandardScaler

df = pd.read_csv('static/wbc.csv') # Mengimport data CSV pada path ke dalam objek DataFrame
df.info() # Menampilkan Informasi Dataframe
df.head() # Menampilkan Tabel dengan 5 data teratas
```

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df1 = df.drop(['id', 'Unnamed: 32'], axis=1) # Menghilangkan kolom dengan nama 'id' dan 'Unnamed: 32'
df1.head()
```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df1['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding merubah nilai kolom diagnosis menjadi numerik
df1.info() # Menampilkan Informasi Dataframe
df1.head() # Menampilkan Tabel dengan 5 data teratas
```

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
all_column = df1.select_dtypes(include=['float64']) # Memilih semua kolom dengan tipe data float64
std = StandardScaler() # Membuat objek StandardScaler
df1[all_column.columns] = std.fit_transform(df1[all_column.columns]) # Melakukan standarisai pada seluruh kolom numerik
df1.head() # Menampilkan Tabel dengan 5 data teratas
```

4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
from sklearn.model_selection import train_test_split # Import library train_test_split

df_train, df_test = train_test_split(df1, test_size=0.2, random_state=0, stratify=df['diagnosis']) # Membagi data df1 dengan presentase df_train 80% dan df_test 20%, dengan perbandingan data di Kolom diagnosis tetap sama antara df_train denga df_test

# Menampilkan jumlah data pada setiap variabel
print(f'Jumlah data asli: {df.shape[0]}')
print(f'Jumlah data latih: {df_train.shape[0]}')
print(f'Jumlah data test: {df_test.shape[0]}')

# Menampilkan Jumlah label data di setiap variabel
print(f'Jumlah label data asli:\n{df1.diagnosis.value_counts()}')
print(f'Jumlah label data latih:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data uji:\n{df_test.diagnosis.value_counts()}')

# df_train.head(10)
# df_test.head(10)
```
